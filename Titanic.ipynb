```python
import pandas as pd
import numpy as np
```

```python
#Import File
df = pd.read_csv("titanic.csv")
```

```python
df.head(10)
```

```python
df.info()
```

```python
any(df.duplicated())
```

```python
#Duplicate Check
any(df.duplicated(['Embarked','Ticket']))
```

```python
dt = df[df.duplicated(['Embarked','Ticket'])]
dt[['Name','Embarked','Ticket']].sort_values(by=['Ticket','Embarked']).head()
```

```python
#drop duplicate
df = df.drop_duplicates(['Embarked','Ticket'])
```

```python
df = df.reset_index()
df = df.drop('index',axis=1)
```

```python
df.head()
```

```python
#drop multiple column
df = df.drop(['Embarked','Cabin'],axis=1)
```

```python
df.head()
```

```python
#Split first name and last name
NameSplit = df['Name'].str.split(',', n = 1, expand = True)
```

```python
df['FirstName'] = NameSplit[1]
df['LastName'] = NameSplit[0]
df = df.drop('Name',axis=1)
```

```python
df.head()
```

```python
cols = df.columns.tolist()
cols = [cols[0]] + cols[9:] + cols[1:9]
df = df[cols]
```

```python
df.head()
```

```python
df['FirstName'] = df['FirstName'].str.replace(" ","")
df['LastName'] = df['LastName'].str.replace(" ","")
df['FirstName'] = df['FirstName'].str.lower()
df['LastName'] = df['LastName'].str.lower()
```

```python
df.head()
```

```python
group = df.groupby('LastName').size().reset_index(name='count')
group[group['count'] >= 2].head()
```

```python
df.sort_values(by='LastName').head(15)
```

```python
def cleanTicket(Ticket) :
    Ticket = Ticket[Ticket.find(' ')+1:]
    return Ticket
df['Ticket'] = df['Ticket'].apply(cleanTicket)
```

```python
df.head()
```

```python
#check null data
df[df.isnull().any(axis=1)].head()
```

```python
#impute null data with the mean
mean = df['Age'].mean()
print(np.round(mean))
df['Age'] = df['Age'].fillna(value = np.round(mean))
```

```python
df[df.isnull().any(axis=1)]
```

```python
#drop row with null/Nan data
df = df.dropna()
```
